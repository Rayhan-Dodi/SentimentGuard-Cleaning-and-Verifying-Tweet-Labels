```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# This can be used to load the dataset
df = pd.read_csv(Path().cwd().joinpath("lab_1_tweet_emotions.csv"))
df.head()
```

```python
# Your code goes here
# Clean data
df_clean = df.dropna(subset=['sentiment'])
df_clean = df_clean.drop_duplicates(subset=['tweet_id'])

# Convert to list
tweet_ids = df_clean['tweet_id'].tolist()

# Print as ONE clean line
print(",".join(map(str, tweet_ids)))
```
